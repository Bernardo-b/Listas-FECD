Slide 10

```python
import pandas as pd
import matplotlib.pyplot as plt
from math import log
import numpy as np
import stemgraphic as sg
import seaborn as sns
```

```python
df = pd.read_csv("BloodPressure.txt")

df.shape

df = df[df.columns[:18]]
df.columns
```

Slide 13

```python
fig, ax = plt.subplots(figsize=(10,8))
ax.hist(df.sbp)
ax.plot(df.sbp, [0.01]*len(df.sbp), '|', color='k')
```

Slide 15

```python
fig, axs = plt.subplots(2, 2,figsize=(15,15))
axs[0, 0].hist(df.sbp)
axs[0, 0].set_title('Histogram of sbp')
axs[0, 0].set_ylabel('Frequency')
axs[0, 0].set_xlabel('sbp')
axs[0, 1].hist(df.sbp, density=True)
axs[0, 1].set_title('Histogram of sbp')
axs[0, 1].set_ylabel('Density')
axs[0, 1].set_xlabel('sbp')
axs[1, 0].hist(df.sbp, density=True, bins = [50,130,160,200,max(df.sbp)])
axs[1, 0].set_title('PS')
axs[1, 0].set_ylabel('Density')
axs[1, 0].set_xlabel('sbp')
axs[1, 1].plot(df.sbp, [-0.001]*len(df.sbp), '|', color='black')
axs[1, 1].hist(df.sbp, density=True, bins = 13,zorder=0)
axs[1, 1].set_title('Pressão sistólica')
axs[1, 1].set_xlabel('pressão')
axs[1, 1].set_ylabel('Density')
plt.tight_layout()
```

### Parte 2

- Usando pandas, ler os dados de população por município: arquivo POP2006.csv
- Refazer os histogramas dos slides 17 e 18

```python
df2 = pd.read_csv("POP2006.csv", encoding = "windows-1252")
df2['logscale'] = [log(i,10) for i in list(df2["POP2006"])]
```

Slide 17

```python
fig, axs = plt.subplots(2, 2,figsize=(15,15))
axs[0,0].hist(df2.POP2006)
axs[0,0].set_title('população municipal em 2006')
axs[0,1].hist(df2.POP2006[df2.POP2006<1000000])
axs[0,1].set_title('pop < 1 milhão')
axs[1,0].hist(df2.POP2006[df2.POP2006<500000])
axs[1,0].set_title('pop < 500 mil')
axs[1,1].hist(df2.POP2006[df2.POP2006<100000])
axs[1,1].set_title('pop < 100 mil')
plt.tight_layout()
```

Slide 18

```python
fig = plt.figure(figsize=[15, 12])
logbins = 10**(np.arange(log(min(df2.POP2006),10),log(max(df2.POP2006),10),0.5))
normbins = np.arange(min(df2.logscale),max(df2.logscale),0.5)
ax1 = fig.add_subplot(2, 2, 2)
ax1.hist(df2.POP2006,bins = logbins)
ax1.set_xscale('log')
ax1.set_title('rotulos na escala original')
ax2 = fig.add_subplot(2, 2, 1)
ax2.hist(df2.logscale, bins = normbins)
ax2.set_title('população, escala log_10')
plt.tight_layout()
```
